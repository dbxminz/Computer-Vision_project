```python
import numpy as np, cv2
import matplotlib.pyplot as plt
import cv2
import glob

count = 1
for n in glob.glob('/content/drive/MyDrive/3팀 부장^^/이미지 데이터/FIRE/2.fire_img_generation_fire/*'):
  c = str(count)
  img2 = cv2.imread(n)
  m = '/content/drive/MyDrive/3팀 부장^^/이미지 데이터/FIRE/4.fire_img_rembg_people/'+c+'.png'
  img1 = cv2.imread(m)
  
  img2 = cv2.resize(img2, (400, 600))
  img1 = cv2.resize(img1, (200, 300))

  # 마스크 생성, 합성할 이미지 전체 영역을 255로 세팅

  mask = np.full_like(img1, 200)

  # 합성 대상 좌표 계산
  offset = 50
  height, width = img2.shape[:2]
  center = (200,300) # offset 임의로 조정

  output_dir = '/content/drive/MyDrive/3팀 부장^^/이미지 데이터/FIRE/fire+people(Normal)/'
  output_path = output_dir +c+'.png'

  normal = cv2.seamlessClone(img1, img2, mask, center, cv2.NORMAL_CLONE)
  cv2.imwrite(output_path, normal)

  count += 1
```
