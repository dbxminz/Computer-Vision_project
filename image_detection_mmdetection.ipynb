```python
from mmdet.apis import inference_detector, init_detector, show_result_pyplot

# Choose to use a config and initialize the detector
config = 'configs/yolact/yolact_r101_1x8_coco.py'
# Setup a checkpoint file to load
checkpoint = 'checkpoints/yolact_r101_1x8_coco_20200908-4cbe9101.pth'
# initialize the detector
model = init_detector(config, checkpoint, device='cuda:0')
```

```python
# Use the detector to do inference
count = 1
for i in glob.glob('/content/drive/MyDrive/3팀 부장^^/이미지 데이터/FIRE/8. fire+peope_final/*'):
  c = str(count)
  result = inference_detector(model, i)
  # Let's plot the result
  # classes = ('person')
  output_dir = '/content/drive/MyDrive/3팀 부장^^/이미지 데이터/FIRE/9. fire_peope_detection (8)/'
  output_path = output_dir + c + '.png'

  model.show_result(i, result, out_file=output_path)
  # show_result_pyplot(model, i, result, score_thr=0.3, out_file=output_path)
  # cv2.imwrite(output_path, )

  count += 1
```
