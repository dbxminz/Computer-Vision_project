```python
count = 1
for n in glob.glob('/content/drive/MyDrive/3팀 부장^^/이미지 데이터/FIRE/3.fire_img_generation_people/*.jpg'):
  c = str(count)
  output_dir = '/content/drive/MyDrive/3팀 부장^^/이미지 데이터/FIRE/4/'
  output_path = output_dir + c + '.png'


  f = np.fromfile(n)
  result = remove(f)
  img = Image.open(io.BytesIO(result)).convert("RGBA")
  img.save(output_path)
  
  count += 1
```
